```python
def delete_rts(env):
    start = datetime.datetime.utcnow()
    print(f"Started at {start}")
    size = 0
    ld4p_group = requests.get(f'https://trellis.{env}.sinopia.io/repository/ld4p')
    ld4p_graph = rdflib.ConjunctiveGraph()
    ld4p_graph.parse(data=ld4p_group.text, format='turtle')
    for obj in ld4p_graph.objects(predicate=LDP.contains):
        print(obj, end="")
        delete_request = requests.delete(str(obj), headers=headers)
        print(f"\t{delete_request.status_code}")
        if delete_request.status_code > 300:
            print("Error")
        size += 1
    end = datetime.datetime.utcnow()
    print(f"Finished at {end} total time {(end-start).seconds} for {size}")
    return ld4p_graph
```

```python
def get_resources(graph):
    rdf_resources = []
    for obj in graph.objects(predicate=LDP.contains):
        if str(obj).endswith("ld4p"):
            continue
        print(obj)
        group_result = requests.get(str(obj))
        group_graph = rdflib.ConjunctiveGraph()
        group_graph.parse(data=group_result.text, format='turtle')
        for resource in group_graph.objects(predicate=rdflib.URIRef('http://www.w3.org/ns/ldp#contains')):
            rdf_resources.append(resource)
            print(".", end="")
    return rdf_resources
```
