```python
import datetime
import rdflib
import json
import requests

now = datetime.datetime.utcnow()

with open("/Users/jpnelson/2020/ld4p/sinopia_sample_profiles/profiles/v0.1.0/SchemaCreativeWork.json") as fo:
    SchemaCreativeWork = json.load(fo)
```

```python
SchemaCreativeWork['Profile']['resourceTemplates'][0]['propertyTemplates']
```

```python
schema_request = requests.get('https://schema.org/CreativeWork.ttl')
schema_org_creative_work = rdflib.ConjunctiveGraph()
schema_org_creative_work.parse(data=schema_request.text, format='turtle')
```

```python
len(schema_org_creative_work)
```

```python
print(schema_org_creative_work.serialize(format='turtle').decode())
```

```python
query = schema_org_creative_work.query("""PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> 
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX schema: <http://schema.org/>

SELECT ?prop_uri ?prop_label
{
  ?prop_uri a rdf:Property .
  ?prop_uri schema:domainIncludes schema:CreativeWork .
  ?prop_uri rdfs:label ?prop_label .
}""")
```

```python

for row in query:
    print(row)
    break
```

```python
print(f"Started at {now} Initial size {len(SchemaCreativeWork['Profile']['resourceTemplates'][0]['propertyTemplates'])}")
for row in query:
    SchemaCreativeWork['Profile']['resourceTemplates'][0]['propertyTemplates'].append({
     'mandatory': 'false',
     'repeatable': 'true',
     'type': 'literal',
     'propertyURI': f"{row[0]}",
     'propertyLabel': f"{row[1]}"   
    })
print(f"Finished size of properties {len(SchemaCreativeWork['Profile']['resourceTemplates'][0]['propertyTemplates'])}")
```

```python
with open("/Users/jpnelson/2020/ld4p/sinopia_sample_profiles/profiles/v0.1.0/SchemaCreativeWork.json", "w+") as fo:
    json.dump(SchemaCreativeWork, fo, indent=2, sort_keys=True)
```
