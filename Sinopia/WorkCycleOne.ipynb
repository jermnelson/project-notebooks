```python
editor_closed_first100 = github.issues_query(repo='sinopia_editor', state='CLOSED')
```

```python
import json
with open("../data/json/issues-closed-001.json", "w+") as fo:
    json.dump(editor_closed_first100, fo, indent=2, sort_keys=True)
    
```

```python
closed, created, title = {}, {}, {}
for issue in editor_closed_first100['data']['repository']['issues']['edges']:
    node = issue['node']
    created[node['number']] = pd.Timestamp(node['createdAt'])
    closed[node['number']] = pd.Timestamp(node['closedAt'])
    title[node['number']] = node['title']
    
closed100_df = pd.DataFrame({ 'created': created, 'closed': closed, 'title': title})

closed100_df['elapsed'] = closed100_df['closed'] - closed100_df['created'] 
```
